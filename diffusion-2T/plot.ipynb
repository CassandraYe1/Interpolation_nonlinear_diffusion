```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
notebook_dir = os.path.dirname(os.path.abspath('.'))
```

zconst-const:

$z=1$

$\beta(x,y,t) = \max\{20t, 10\}, g(x,y,t) = 0.01$

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 4), dpi=200)
img_path = os.path.join(notebook_dir, 'diffusion-2T/results/zconst-const/')
img_E = mpimg.imread(os.path.join(img_path, 'fig_E.png'))
img_T = mpimg.imread(os.path.join(img_path, 'fig_T.png'))
ax1.imshow(img_E)
ax1.axis('off')
ax2.imshow(img_T)
ax2.axis('off')
plt.tight_layout()
plt.show()
```
